```python
import tensorflow as tf

x_data = [1,2,3]
y_data = [1,2,3]

W = tf.Variable(tf.random_normal([1]), name='weight')
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

hypothesis = W*X

cost = tf.reduce_mean(tf.square(hypothesis - Y))

# 수동으로 cost를 minimize하는 과정
learning_rate = 0.1
# gradient는 '경사', 경사는 미분해서 구함, 그래서 미분한 식을 넣어줌
gradient = tf.reduce_mean((W*X - Y)*X)
descent = W - learning_rate * gradient
update = W.assign(descent) # session(세션) 속에서 값을 변경하고 싶을 땐 assign

sess = tf.compat.v1.Session()
sess.run(tf.global_variables_initializer()) # 변수 초기화
for i in range(21):
    # 텐서로 만들어놓고 실행은 여기서 시키는 것 같다
    sess.run(update, feed_dict={X:x_data,Y: y_data})
    print(i, sess.run(cost, feed_dict={X: x_data, Y: y_data}),sess.run(W))
```

```python
import tensorflow as tf

# 오른쪽에 있는 5.0을 줘서 기울기가 양수니까 x를 음수쪽으로 가게 함
W = tf.Variable(5.0) # 말도 안되는 멀리있는 값을 일부러 줌, 보기위해서
X = [1,2,3]
Y = [1,2,3]

hypothesis = W*X

cost = tf.reduce_mean(tf.square(hypothesis - Y))

########## 자동 미분해서 gradient_Descent하는 법
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1)
train = optimizer.minimize(cost) # cost를 굳이 미분하지 않아도 자동으로 됨
###############
sess = tf.compat.v1.Session()
sess.run(tf.global_variables_initializer()) # 변수 초기화
for i in range(100):
    print(i,sess.run(W))
    sess.run(train)
```

```python
# Gradient 값 수정하는 법 ( 손 보는 법 )
gvs = optimizer.compute_gradients(cost)
# ... ( 손 봄 ) 
apply_gradients = optimizer.apply_gradients(gvs)
```
