```python
# 상수가 아닌 실행할 때 값을 주게 하기
import tensorflow as tf
# ->tf.placeholder 사용
X = tf.placeholder(tf.float32, shape = [None]) # None : 몇차원인지 쓰는것, none이면 아무거나 가능
Y = tf.placeholder(tf.float32)

W = tf.Variable(tf.random_normal([1]), name='weight') # 첫 매개변수는 몇차원인지 알려주는것
b = tf.Variable(tf.random_normal([1]), name='bias')

# H(x) = Wx + b
hypothesis = X * W + b

#cost function 계산식 적용
cost = tf.reduce_mean(tf.square(hypothesis - Y)) #reduce_mean : 평균내주는거

# GradientDescent
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost) # variable을 조정해서 가장 작은거 찾음

# 그래프를 세션에서 실행
sess = tf.compat.v1.Session()
# W,b 라는 변수를 사용했는데, 이걸 사용할 땐 tf.global_variables_initializer를 해야함
sess.run(tf.global_variables_initializer())

for step in range(2001) :
     # _ : 필요없어서 return 안받겠다, \ : 코드가 길어서 밑에 쓰겠다
    cost_val, W_val, b_val, _ = \
    sess.run([cost, W, b, train], feed_dict={X:[1,2,3],Y:[1,2,3]}) #sess.run(cost),sess.run(W) ... 이걸 리스트로 묶어서 한번에 가능
    if step % 20 == 0 : # 과정을 눈으로 보기 위함
        print(step, cost_val, W_val, b_val)
    
```
